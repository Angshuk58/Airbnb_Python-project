```python
# Import the librarires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the dataset
path = '/content/drive/MyDrive/Data Science/AirBnB Project/AirBnB Project/Airbnb NYC 2019.csv'
df = pd.read_csv(path)
df.head()
```

```python
# Description of the dataset
df.describe()
```

```python
# Info of the dataset
df.info()
```

```python
# Need to find rows with null value
df.isnull().sum()
```

```python
# Remove the rows having name as null
df = df[df['name'].notnull()]
```

```python
df.isnull().sum()
```

```python
# remove the rows where host_name is null
df = df[df['host_name'].notnull()]
```

```python
df.isnull().sum()
```

```python
# Find Distribution Of Airbnb Bookings Price Range?
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.histplot(df['price'])
plt.subplot(1,2,2)
sns.boxplot(df['price'])

```

```python
# Analysis of the data
# Using bar plot to find the price distribution using roomtype as hue
sns.barplot(x='room_type', y='price', hue='room_type', data=df)
```

```python
# Find Distribution Of Airbnb Bookings Price Range?
df['price'].hist()

```

```python
# Using bar plot to find the total listing in each neighbourhood group using hue as different groups
sns.countplot(x='neighbourhood_group', hue='room_type', data=df)

```

```python
#average price of listing/property in each neighbourhood group
avg_ng=df.groupby('neighbourhood_group')['price'].mean()
avg_ng

```

```python
# Barplot of avg_ng with neighbourhood_group
sns.barplot(x=avg_ng.index, y=avg_ng.values, palette='mako')
```

```python
#average price of listing/property in each neighbourhood
avg_neighbour=df.groupby('neighbourhood')['price'].mean()
avg_neighbour
```

```python
# Barplot of avg_neighbour with neighbourhood
sns.barplot(x=avg_neighbour.index, y=avg_neighbour.values, palette='mako')
```

```python
# top 50 neighbourhoods and host by listing/property in entire NYC
df.groupby(['neighbourhood','host_name']).size().sort_values(ascending=False).head(50)

```

```python
# barplot of Number of active Hosts per location by each neighbourhood group
sns.countplot(x='neighbourhood_group', palette='mako', data=df)


```

```python
# Total counts of each room types in NYC
sns.countplot(x='room_type', palette='mako', data=df)

```

```python
# stay requirement counts by minimum nights
sns.countplot(x='minimum_nights', palette='mako', data=df)
```

```python
stay_requirements = df['minimum_nights'].value_counts()
print(stay_requirements)
```

```python
# the total numbers of reviews and Maximum reviews by each neighbourhood group
total_reviews=df.groupby('neighbourhood_group')['number_of_reviews'].sum()
total_reviews
```

```python
# Barplot of total_reviews with neighbourhood_group
sns.barplot(x=total_reviews.index, y=total_reviews.values, palette='mako')
```

```python
# most reviewed room type in neighbourhood groups per month
most_reviewed_room=df.groupby(['neighbourhood_group','room_type']).size().sort_values(ascending=False)
most_reviewed_room
```

```python
# Barplot of most_reviewed_room with neighbourhood group
sns.barplot(x=most_reviewed_room.index.get_level_values(0),
            y=most_reviewed_room.values,
            palette='mako')
```

```python
# best location listing/property location for travelers
best_listings=df.groupby('neighbourhood').size().sort_values(ascending=False)
best_listings
```

```python
# Barplot of best_listings
sns.barplot(x=best_listings.index, y=best_listings.values, palette='mako')
```

```python
# best location listing/property locations for hosts
df.groupby('neighbourhood')['availability_365'].sum().sort_values(ascending=True)
```

```python
# price variations in NYC neighbourhood groups
price_variation=df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
price_variation
```

```python
# Barplot of price_variation with neighbourhood_group
sns.barplot(x=price_variation.index, y=price_variation.values, palette='mako')
```

```python
# Map based on Neighbourhood group
plt.figure(figsize=(10,6))
sns.scatterplot(x=df.longitude, y=df.latitude, hue=df.neighbourhood_group) # Pass x and y as keyword arguments
plt.ioff()
```

```python
# Map based on room type
plt.figure(figsize=(10,6))
sns.scatterplot(x=df.longitude, y=df.latitude, hue=df.room_type) # Pass x and y as keyword arguments
plt.ioff()
```
